````python

class llm_chain:
    def __init__(self, prompt_template):
        prompt = prompt_template
        # memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True, input_key="question")
        self.chain = LLMChain(llm=llm, prompt=prompt, verbose=True)
    
    def run(self, inp):
        # response = self.chain.run({question: question, name: name})
        # time = get_current_time()
        # print("hitting llm ",time)
        response = self.chain.run(inp)

        # time = get_current_time()
        # print("done ",time)

        return response

        
````

````python
prompt_messages = [
    SystemMessagePromptTemplate.from_template(INTENT_SENTIMENT_CLASSIFICATION_PROMPT["SYSTEM"]),
    HumanMessagePromptTemplate.from_template(INTENT_SENTIMENT_CLASSIFICATION_PROMPT["CONTEXT"])
]

email_content_template =  """
Email Thread Summary: {summary}

Current email content: {question}
"""
prompt_messages.append(HumanMessagePromptTemplate.from_template(email_content_template))

prompt_messages.append(SystemMessagePromptTemplate.from_template(INTENT_SENTIMENT_CLASSIFICATION_PROMPT["DISPLAY"]))
prompt_messages.append(SystemMessagePromptTemplate.from_template(INTENT_SENTIMENT_CLASSIFICATION_PROMPT["REMEMBER"]))
chat_prompt = ChatPromptTemplate.from_messages(prompt_messages)

classification_prompt_chain = llm_chain(chat_prompt)
````
